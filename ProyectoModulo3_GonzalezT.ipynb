# Marco te√≥rico 
Para entender el problema, es necesario abordar los siguientes conceptos:

* Hidrogeles Superabsorbentes (SAPs): Son redes tridimensionales de pol√≠meros hidrof√≠licos capaces de absorber grandes cantidades de agua. Donde sus propiedades como el grado de hinchamiento y la estabilidad mec√°nica dependen de la reticulaci√≥n y composici√≥n qu√≠mica. Sus propiedades clave son: grado de hinchamiento, estabilidad mec√°nica y biodegradabilidad.

* Biocarb√≥n (Biochar): Es un material obtenido por pir√≥lisis (la degradaci√≥n de la biomasa por medio del efecto del calor sin que se necesite la presencia de ox√≠geno) de biomasa en condiciones de bajo ox√≠geno. Mejora la retenci√≥n de agua y nutrientes en suelos, adem√°s de aumentar la estabilidad del hidrogel.
  
* Proceso de S√≠ntesis del Hidrogel:
Se le agreg√° el prepol√≠mero (El sorbitol y el √°cido c√≠trico se mezclan en una relaci√≥n fija de 1:5), el CMC, el biocarb√≥n y el agua destilada. Se mezcla y se somete a un tratamiento t√©rmico el cual es el proceso de entrecruzamiento para obtener una estructura estable. Posteriormente se tritura y se tamiza.

* Capacidad de Absorci√≥n Libre (FSC): Es una prueba para medir la cantidad de agua retenida por unidad de masa del hidrogel, en este caso en soluci√≥n salina al 0.12% siendo representativa de las aguas de riego ,siendo un indicador clave de la funcionalidad del material.
